```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['gender'].value_counts()
```

```python
df['gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Observation:

This bar chart shows  that the number of male students and is higher than female students.

insight:
This indicates that male students are more represented in the dataset compared to female students.


```python
df['StudentAbsenceDays'].value_counts()
```

```python
df['StudentAbsenceDays'].value_counts().plot(kind='bar')
plt.title('Student Absence Days')
plt.xlabel('Absence Category')
plt.ylabel('Number of Students')
plt.show()
```

Observation:
The bar charts shows tha the number of students with under7 absence days is higher than those with above7 absence days

insight:
This indicates that most students attend classes regularly and have low absenteeis

```python
df['GradeID'].value_counts()
```

```python
df['GradeID'].value_counts().plot(kind='bar')
plt.title('Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()
```

Observation:
The chart shows that Grade G-02 has the highest number of students, followed by G-08 and G-07, while higher grades have fewer students.

insight:
Lower grades have more students enrollment compared to higher grades in this dataset.

```python
sns.countplot(x='gender', hue='StudentAbsenceDays',data=df)
plt.title('Gender vs Students Absence')
plt.show()
```

Observation:
For both male and female students, the number of students with under7 absence days is higher than those with above7 absence days.

insight:
Attendance patterns are similar for both genders, and most students from both groups shows regular attendance.

Conclusion:

This project analyzed student demographic and attendance data using data visualization techniques. The results show higher male participation, regular attendance among most students, and higher enrollment in lower grades. Overall, attendance patterns are similar across genders.
